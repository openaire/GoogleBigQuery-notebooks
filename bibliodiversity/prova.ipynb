```python
%%bigquery
WITH products as (
  SELECT id, languageCode FROM oag_v10_6_0.publications
  UNION ALL
  SELECT id, languageCode FROM oag_v10_6_0.others
  UNION ALL
  SELECT id, languageCode FROM oag_v10_6_0.datasets
  UNION ALL
  SELECT id, languageCode FROM oag_v10_6_0.software
)
SELECT languageCode, COUNT( DISTINCT id) AS productCount
FROM products
  WHERE languageCode != "und"
GROUP BY languageCode
ORDER BY productCount DESC
```
