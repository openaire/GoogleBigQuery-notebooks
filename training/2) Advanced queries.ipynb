# **Task:** count citations of publications

```python
%%bigquery

SELECT cited.id, COUNT(*) as citations
FROM oag_v9_0_1.publications as cited
  JOIN oag_v9_0_1.relations as rel
    ON cited.id = rel.target
  JOIN oag_v9_0_1.publications as citing
    ON citing.id = rel.source
WHERE rel.relationName = 'Cites'
GROUP BY cited.id
ORDER BY citations DESC
LIMIT 100
```

## **Subtask:** count citations of datasets

```python
%%bigquery

SELECT cited.id, COUNT(*) as citations
FROM oag_v9_0_1.datasets as cited
  JOIN oag_v9_0_1.relations as rel
    ON cited.id = rel.target
  JOIN oag_v9_0_1.publications as citing
    ON citing.id = rel.source
WHERE rel.relationName = 'Cites'
GROUP BY cited.id
ORDER BY citations DESC
LIMIT 100
```

# **Task:** count publications by country (of affiliated organisations)

```python
%%bigquery

SELECT countryCode, COUNT(DISTINCT publications.id) AS n_pubs
FROM oag_v9_0_1.publications,
  oag_v9_0_1.relations,
  oag_v9_0_1.organizations
WHERE publications.id = relations.source AND
      relations.target = organizations.id AND
      relations.relationName = 'hasAuthorInstitution'
GROUP BY countryCode
ORDER BY n_pubs DESC
```

# **Task:** count projects by country (of participating organisations)

```python
%%bigquery

SELECT orgs.countryCode AS country, COUNT(proj.id) as n_projects
FROM oag_v9_0_1.organizations AS orgs,
    oag_v9_0_1.projects AS proj,
    oag_v9_0_1.relations AS rels
WHERE orgs.id = rels.source AND
        proj.id = rels.target
GROUP BY country
ORDER BY n_projects DESC
```

# **Task:** find the projects with most research products associated

```python
%%bigquery

SELECT proj.id, proj.title, COUNT(pubs.id) as n_products
FROM oag_v9_0_1.projects as proj,
      oag_v9_0_1.relations as rels,
      oag_v9_0_1.publications as pubs
WHERE proj.id = rels.source AND
      pubs.id = rels.target
GROUP BY proj.id, proj.title
ORDER BY n_products DESC
LIMIT 25
```

`unidentified` is a cheatcode to associate products to founders when the funder is known while the project is not.

It could be filtered. Can you do that?

Also, in the previous query we accounted for publications only. If we want to count products other than literature, we can use a `UNION`

```python
%%bigquery

WITH all_products AS (
  SELECT id FROM oag_v9_0_1.publications
  UNION ALL
  SELECT id FROM oag_v9_0_1.datasets
  UNION ALL
  SELECT id FROM oag_v9_0_1.software
  UNION ALL
  SELECT id FROM oag_v9_0_1.others
)
SELECT proj.id, proj.title, COUNT(all_products.id) as n_products
FROM oag_v9_0_1.projects as proj,
      oag_v9_0_1.relations as rels,
      all_products
WHERE proj.id = rels.source AND
      all_products.id = rels.target AND
      proj.title != 'unidentified'
GROUP BY proj.id, proj.title
ORDER BY n_products DESC
LIMIT 25
```

# **Task:** find most prolific author collaborations

```python
%%bigquery

WITH paper_author AS (
  SELECT id as paper_id, JSON_VALUE(author.pid.id.value) AS fullname
    FROM oag_v9_0_1.publications,
      UNNEST(JSON_QUERY_ARRAY(publications.authors)) AS author
    -- WHERE EXTRACT(YEAR FROM DATE(publicationDate)) = 2024 AND
      WHERE JSON_VALUE(author.pid.id.scheme) = 'orcid'
)

SELECT l.fullname, r.fullname, COUNT(*) as n_papers
FROM paper_author AS l
  JOIN paper_author AS r
ON l.paper_id = r.paper_id
WHERE l.fullname != r.fullname AND l.fullname < r.fullname
GROUP BY l.fullname, r.fullname
ORDER BY n_papers DESC
```

# **Task:** find countries collaborating by mean of projects

```python
%%bigquery

WITH country_project AS (
  SELECT countryCode AS country,
           target AS project_id
  FROM oag_v9_0_1.organizations JOIN oag_v9_0_1.relations
      ON relationName = 'isParticipant' AND source = organizations.id
  WHERE country IS NOT NULL
)

SELECT l.country AS left_country,
       r.country AS right_country,
       COUNT(*) AS count
FROM country_project AS l JOIN country_project AS r
      ON l.project_id = r.project_id AND l.country <= r.country
GROUP BY left_country, right_country
ORDER BY count DESC
```
